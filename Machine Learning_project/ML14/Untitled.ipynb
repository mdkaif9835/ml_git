```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_csv('titanic.csv')
df.head()
```

```python
rd= df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis='columns')
rd.head()
```

```python
dummies = pd.get_dummies(rd.Sex)
dummies.head()
```

```python
rd = pd.concat([rd,dummies],axis='columns')
rd.head()
```

```python
rd.drop(['Sex'], axis = 'columns',inplace=True)
rd.head()
```

```python
target=rd.Survived
target.head()
```

```python
x= rd.drop('Survived',axis='columns')
x.head()
```

```python
x.columns[x.isna().any()]
```

```python
x.Age[0:30]
```

```python
x.Age=x.Age.fillna(x.Age.mean())
x.head(12)
```

```python
x.Fare=x.Fare.fillna(x.Fare.mean())
x.head(30)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,target, test_size=0.2)
```

```python
len(x_train)
```

```python
len(x)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```
