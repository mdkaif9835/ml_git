```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
fr = pd.read_csv('tested.csv')
fr.head()
```

```python
psd= fr.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'], axis='columns')
psd.head()
```

```python
dummy = pd.get_dummies(psd.Sex)
dummy.head()
```

```python
merge= pd.concat([psd,dummy], axis='columns')
merge
```

```python
titf= merge.drop(['Sex'], axis='columns')
titf
```

```python
import math
Median_Exp = math.floor(titf.Age.mean())
Median_Exp
```

```python
c=3
c
```

```python
s=0
s
```

```python
f=10
f
```

```python
titf.Age= titf.Age.fillna(Median_Exp)
titf.Survived= titf.Survived.fillna(s)
titf.Pclass= titf.Pclass.fillna(c)
titf.Fare= titf.Fare.fillna(f)
titf
```

```python
x= titf.drop(['Survived'], axis='columns')
x
```

```python
target= titf.Survived
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
LabelEncoder()
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(x,target)
```

```python
model.score(x,target)
```
